```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.backends.backend_pdf



all_sheets = pd.read_excel('inventaire 23-24.xlsx', sheet_name=None)
df = pd.concat(all_sheets.values(), ignore_index=True)

#pdf
```

```python
fig = plt.figure(figsize=(12, 6))

# pour une meilleure lisibilit√©.
sns.set(style="whitegrid")
# taille globale du graphique.
plt.figure(figsize=(16, 8))

#cat√©gories les plus fr√©quentes dans la colonne "D√©partement/Services"

top_categories = df['D√©partement/Services'].value_counts().nlargest(10).index
df_filtered = df[df['D√©partement/Services'].isin(top_categories)]
plt.figure(figsize=(16, 8))
#diagramme √† barres 
sns.countplot(x="D√©partement/Services", data=df_filtered)

# ajout des √©tiquettes
plt.title("Nombre de mat√©riels par les 10 d√©partements/services les plus fr√©quents", fontsize=16)
plt.xlabel("D√©partement/Service", fontsize=14)
plt.ylabel("Nombre de mat√©riels", fontsize=14)

# les √©tiquettes de l'axe x √† la verticale 
plt.xticks(rotation=90, ha="right")

# Enregistrement 
#plt.savefig('nbMatosLesPLUSFREQUENT.png', bbox_inches='tight')

```

```python
fig = plt.figure(figsize=(12, 6))
top_brands = df['Marque'].value_counts().nlargest(10).index
df_filtered_brands = df[df['Marque'].isin(top_brands)]

# Diagramme circulaire pour la r√©partition des 10 marques les plus fr√©quentes
plt.figure(figsize=(10, 10))
df_filtered_brands['Marque'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90)
plt.title("R√©partition des 10 Marques les plus fr√©quentes")
plt.show()
```
